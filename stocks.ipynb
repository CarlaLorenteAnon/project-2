This script takes historical stock data for H&M 

```python
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
```

```python
# https://www.nasdaq.com/market-activity/stocks/hnnmy/historical

stocks_df = pd.read_csv('Resources/stocks.csv')

stocks_df
```

```python
# set start and end date

start_date = '09/19/2018'
start_date = dt.strptime(start_date, '%m/%d/%Y')

end_date = '09/21/2020'
end_date = dt.strptime(end_date, '%m/%d/%Y')
```

```python
# filter dates to match transaction data

stocks_df = stocks_df[(stocks_df['Date'].apply(lambda x: dt.strptime(x, '%m/%d/%Y')) <= end_date) & 
        (stocks_df['Date'].apply(lambda x: dt.strptime(x, '%m/%d/%Y')) >= start_date)]
stocks_df
```

```python
stocks_df.plot(kind='line', x='Date');
```
